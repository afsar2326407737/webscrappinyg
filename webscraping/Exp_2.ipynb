```python
df.drop(['Title','Tags','View Rating','Release Date','Netflix Release Date','Production House','Netflix Link','IMDb Link','Metacritic Score','Boxoffice',
         'Summary','Poster','TMDb Trailer','Trailer Site'],axis=1,inplace=True)
```

```python
df.shape
```

```python
df.dropna(subset = ['Director'],inplace=True)
df.dropna(subset = ['Writer'], inplace = True)
df.dropna(subset = ['Actors'], inplace = True)
df.dropna(subset = ['Runtime'], inplace = True)
df.dropna(subset = ['Genre'], inplace = True)
# df.dropna(subset = ['Country Avalability'], inplace = True)
# df.dropna(subset = ['Language'], inplace = True)
```

```python
sns.heatmap(df.isnull())
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df['Hidden Gem Score'].ffill(axis=0 , inplace=True)
df['IMDb Score'].bfill(axis=0 , inplace = True)
```

```python
df.isna().sum()
```

```python
df['IMDb Votes'] = df['IMDb Votes'].fillna(df['IMDb Votes'].mean(),inplace = True)
```

```python
df.isna().sum()
```

```python
df['Awards Nominated For'] = df['Awards Nominated For'].replace(np.nan,0) 
df['Awards Received'] = df['Awards Received'].fillna(0)
```

```python
df.isna().sum()
```

```python
df
```

```python
data = pd.concat([df,pd.get_dummies(df['Series or Movie'])],axis=1)
```

```python
data
```

```python
data.rename(columns = {'Series or Movie' : 'Series / Movie'} , inplace = True)
```
